```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = pd.read_csv('outbreak.csv')
```

```python
df
```

```python
plt.hist(df['State'], bins=10)
plt.title('Frequency of Illness by State')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.show()
```

```python
top5_instances_genus = df['Genus'].value_counts().head(5)

plt.bar(top5_instances_genus.index, top5_instances_genus.values)
plt.title('5 Most Commonly Occurring Bacteria in Food Outbreaks')
plt.xlabel('Bacteria')
plt.ylabel('Frequency')
plt.show()
```

```python
top5_instances_state = df['State'].value_counts().head(5)

plt.bar(top5_instances_state.index, top5_instances_state.values)
plt.title('5 States with the Highest Frequency of Outbreak')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['Year'], bins=10)
plt.title('Years with the Most Outbreaks')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['Month'], bins=10)
plt.title('Months with the Most Outbreaks')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.show()
```

```python
top_instances_food = df['Food'].value_counts().head(3)

plt.bar(top_instances_food.index, top_instances_food.values)
plt.title('Top Foods Associated with Food Borne Illness')
plt.xlabel('Food')
plt.ylabel('Frequency')
plt.show()
```

```python
year_df = df.groupby('Year')['Illnesses'].mean()

plt.plot(year_df.index, year_df.values)
plt.title('Frequency of Illness by State')
plt.xlabel('State')
plt.ylabel('Average Number of Illnesses')
plt.show()
```

```python
grouped_df = df.groupby('Month')['Illnesses'].mean()

plt.plot(grouped_df.index, grouped_df.values)
plt.title('Frequency of Illness in Each Month')
plt.xlabel('Month')
plt.ylabel('Average Number of Illnesses')
plt.show()
```
